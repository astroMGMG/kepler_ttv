```julia
using PyPlot
```

## Original, basic code

The oringal curve_fitting.jl file. The plotting function has been commented out to simplify things, and as we are only looking at run time. This code reads in an ASCII file and will output a fitted light curve, which will be written to a file using the functions below.

```julia
include("curve_fitting_old.jl")
```

We will need to include two functions that aren't in our original curve fitting program, one to write an output file in ASCII, and another to run the code multiple times at once. The code will read in a large (~32 Mb) ASCII data file and output an ASCII data file.

```julia
function multiple(n::Int)
    for j in 1:n
        file = fit_curve("test_csv_huge_$j.csv")
        write_file(file,"output_$j.csv")
    end
end
```

```julia
function write_file(x::Array,name)
    #Open file for writing. x Should be an array of real numbers. File format is ASCII
    filen=open(name,"w")
    writecsv(filen,x)
    close(filen)
end
```

```julia
time_csv_orig_1=@elapsed multiple(1);
time_csv_orig_3=@elapsed multiple(3);
time_csv_orig_5=@elapsed multiple(5);
time_csv_orig_7=@elapsed multiple(7);
time_csv_orig_10=@elapsed multiple(10);
```

## Optimized code

This is the sped up file that is using push! instead of concating variables as width=[width,test[i]]. Additionally, this file uses binary files instead of csv files.

```julia
include("curve_fitting_speed.jl")
```

We also need a write file for binary files now, and we need to redefine the multiple function to take in binary files.

```julia
function write_file_binary(x::Array,name)
    #Write an array of real numbers to a file. Format is binary
    filen=open(name,"w")
    write(filen,x)
    close(filen)
end
```

```julia
function multiple(n::Int)
    for j in 1:n
        file = fit_curve("test_huge_binary_$j.dat")
        write_file(file,"output_$j.dat")
    end
end
```

```julia
time_binary_speed_1=@elapsed multiple(1);
time_binary_speed_3=@elapsed multiple(3);
time_binary_speed_5=@elapsed multiple(5);
time_binary_speed_7=@elapsed multiple(7);
time_binary_speed_10=@elapsed multiple(10);
```

## Comparison

```julia
files=[1,3,5,7,10];
original_time=[time_csv_orig_1,time_csv_orig_3,time_csv_orig_5,time_csv_orig_7,time_csv_orig_10];
speed_time=[time_binary_speed_1,time_binary_speed_3,time_binary_speed_5,time_binary_speed_7,time_binary_speed_10];

p1=plot(files,log10(original_time))
p2=plot(files,log10(speed_time))

legend([p1,p2],["Original code","Optimized code"],loc=2)
plt.xlabel("Number of Files")
plt.ylabel("log(Time (seconds))")
```

We see a huge improvement in run time with the sped up code, using push! and binary files. In fact, we need to take the log of the time just to see both curves on the same plot. The run time here varies linearly with number of files. We can now look at parallelizing the code.

# Parallelization

## On Desktop

Here, @spawn statements have been added for finding the width, depth and transit times. The tasks will then be turned over to any free processors.

```julia
include("curve_fitting_parallel.jl")
```

Lets look at the run time for different processors reading in 10 files

```julia
time_parallel_1=@elapsed multiple(10);
addprocs(1);
@everywhere include("curve_fitting_parallel.jl");
time_parallel_2=@elapsed multiple(10);
addprocs(1);
@everywhere include("curve_fitting_parallel.jl");
time_parallel_3=@elapsed multiple(10);
addprocs(1);
@everywhere include("curve_fitting_parallel.jl");
time_parallel_4=@elapsed multiple(10);
```

```julia
nprocs()
```

```julia
procs=[1,2,3,4];
time_p=[time_parallel_1,time_parallel_2,time_parallel_3,time_parallel_4];
plot(procs,time_p)
plt.xlabel("Number of Processors")
plt.ylabel("Time (seconds)")
```

The time it takes to read 10 files decreases with additional processors, although the time increase is minimal.

## RCC Cluster

The next step is to put it on the RCC cluster. For this, we decided to try and parallelize the "multiple" function as well, by spawning the fit_curve across free processors. However, we did not see any improvement in the RCC cluster, whether or not processors were on multiple nodes. Perhaps the @spawn statements in the "multiple" function may have conflicted with the @spawn statements in the curve fitting function itself, but we're not really sure why we didn't see an improvement. Results are shown below.

```julia
multiple_nodes_time=[6.28666,6.13518,7.84943,8.14873,4.84381];
multiple_nodes_cores=[1,2,4,6,8];

single_node_time=[6.28666,6.82104,7.89302,5.62365,6.62678]
single_node_cores=[1,2,4,6,8];

p1=plot(multiple_nodes_cores,multiple_nodes_time)
p2=plot(single_node_cores,single_node_time,color="red")
legend([p1,p2],["Multiple nodes","Single node"],loc=3)
plt.xlabel("Number of Processors")
plt.ylabel("Time (seconds)")
```

In all, the largest speed up was gained from using the push! command, as well as writing and reading binary files instead of ASCII. Parallelization helped, but if we're running this on the RCC cluster, it will take longer to wait in the queue than it will to run the serial version.
