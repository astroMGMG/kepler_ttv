## One final time . . . (THIS ONE ACTUALLY WORKS!)

```julia
include("get_good_indices.jl")
using PyCall;
@pyimport pyfits;
@pyimport kplr;
#setup link to MAST
client = kplr.API()
```

```julia
koi = client[:koi](952.01)
```

```julia
#print(koi[:koi_period]," ",koi[:koi_period_err1]," ",koi[:koi_period_err2])
```

```julia
lcs = koi[:get_light_curves](short_cadence=false)
```

```julia
time,flux,ferr,quality = [],[],[],[];
for lc in lcs
    hdu = lc[:open]();
    hdu_primary = hdu[1];
    hdu_bintable = hdu[2];
    hdu_image = hdu[3];
    #println(lc)
    #println(hdu)
    #println(hdu_bintable[:data])
    hdu_bintable_data=hdu_bintable[:data];
    blah=keys(hdu_bintable_data)
    timetemp=get(hdu_bintable_data, "time");
    timetemp=convert(Array{Float64,1}, timetemp);
    
    fluxtemp=get(hdu_bintable_data, "pdcsap_flux");
    fluxtemp=convert(Array{Float64,1}, fluxtemp);
    
    #strip out bad data
    ind=get_good_indices(fluxtemp);
    timetemp=timetemp[ind];
    fluxtemp=fluxtemp[ind];
    fluxtemp=fluxtemp/mean(fluxtemp)
    
    time=[time,timetemp];
    flux=[flux,fluxtemp];
    #println(length(time)," ",length(flux))
    #println(hdu_bintable_data["time"])
    #help(hduprimary)
end
```
