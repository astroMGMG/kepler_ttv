# Accessing Kepler Data and Plotting

### HOW TO DO IT IN PYTHON...

####Import functions and load packages as well as establish a link with MAST

```julia
include("get_good_indices.jl")
using PyPlot;
using PyCall;
@pyimport pyfits;
@pyimport kplr;
#setup link to MAST
client = kplr.API();
```

####Create a `koi` (inherits the methods of `client`)

```julia
koi = client[:koi](952.01)
```

####The `koi` now has many methods that can be accessed

```julia
#print(koi[:koi_period]," ",koi[:koi_period_err1]," ",koi[:koi_period_err2])
```

####Retrieve all the long cadence light curves for the `koi`

```julia
lcs = koi[:get_light_curves](short_cadence=false)
```

```julia
#initialize some empty arrays to be filled
time,flux,ferr,quality = [],[],[],[];

#iterate over each lightcurve
for lc in lcs
    #open the lightcurve file for reading
    hdu = lc[:open]();
    
    #hdu_primary = hdu[1];
    
    #this hdu contains the data
    hdu_bintable = hdu[2];
    
    #hdu_image = hdu[3];
    
    hdu_bintable_data=hdu_bintable[:data];
    
    #retrieve the time
    timetemp=get(hdu_bintable_data, "time");
    timetemp=convert(Array{Float64,1}, timetemp);
    
    #retrieve the flux
    fluxtemp=get(hdu_bintable_data, "pdcsap_flux");
    fluxtemp=convert(Array{Float64,1}, fluxtemp);
    
    #strip out bad data
    ind=get_good_indices(fluxtemp);
    timetemp=timetemp[ind];
    fluxtemp=fluxtemp[ind];
    
    #normalize each quarter
    fluxtemp=fluxtemp/mean(fluxtemp);
    
    #build up the combined arrays
    time=[time,timetemp];
    flux=[flux,fluxtemp];
end
```

```julia
plot(time,flux, marker=".",markersize=2,linewidth=0);
xlabel("Time (in days)")
ylabel("Normalized Flux")

#xlim(150,1000)
ylim(0.99,1.01)
```

The above figure shows the total lightcurve over all availible quarters. This koi has 14 quarters of data. Missing data has been stripped out. Certain quarters appear to have false signals in them and will need to be cotrended so that a flat lightcurve can be obtained.
